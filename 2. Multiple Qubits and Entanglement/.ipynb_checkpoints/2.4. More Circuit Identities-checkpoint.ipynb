```python
from qiskit import QuantumCircuit, Aer, assemble
from qiskit.circuit import Gate
from qiskit.visualization import plot_bloch_multivector, array_to_latex
from math import pi
sim = Aer.get_backend('aer_simulator')

qc = QuantumCircuit(2)
c = 0
t = 1
```

## 1. Making a Controlled-Z from a CNOT

```python
qc.cz(c,t)
qc.draw()
```

```python
qc = QuantumCircuit(2)
# also a controlled-Z
qc.h(t)
qc.cx(c,t)
qc.h(t)
qc.draw()
```

- Qiskit에서는 cz가 직접 호출이 가능하지만, 실제 IBM Q 디바이스에서는 그렇지 않다. IBM Q에서 지원하는 것은 CNOT(=controlled X)뿐이다. 따라서 경우에 따라서는 프로그래머가 직접 Controlled-Y, Controlled-Z, Controlled-H 게이트를 직접 설계해서 줘야한다.
- $HXH = Z$, $HZH = X$, $R_{y, \pi/4}XR_{y, -\pi/4} = H$ 임을 이용한다.

```python
qc = QuantumCircuit(2)
# also a controlled-Z
qc.h(t)
qc.cx(c,t)
qc.h(t)
qc.draw()
```

```python
qc = QuantumCircuit(2)
# a controlled-Y
qc.sdg(t)
qc.cx(c,t)
qc.s(t)
qc.draw()
```

```python
qc = QuantumCircuit(2)
# a controlled-H
qc.x(0)
qc.ry(pi/4,t)
qc.cx(c,t)
qc.ry(-pi/4,t)
qc.draw()
```

```python
qc.save_statevector()

qobj = assemble(qc)
final_state = sim.run(qc).result().get_statevector()
plot_bloch_multivector(final_state)
```

## 2. Swapping Qubits

```python
a = 0
b = 1
```

```python
qc = QuantumCircuit(2)
# swaps states of qubits a and b
qc.swap(a,b)
qc.draw()
```

```python
qc = QuantumCircuit(2)
# swap a 1 from a to b
qc.cx(a,b) # copies 1 from a to b
qc.cx(b,a) # uses the 1 on b to rotate the state of a to 0
qc.cx(a,b) # uses the 1 on a to rotate the state of b to 0
qc.draw()
```

```python
qc.save_unitary()
qobj = assemble(qc)
uni = sim.run(qc).result().get_unitary()
array_to_latex(uni)
```

```python
qc = QuantumCircuit(2)
# swap a 1 from a to b
qc.cx(a,b) # copies 1 from a to b
qc.cx(b,a) # uses the 1 on b to rotate the state of a to 0
qc.cx(b,a) # uses the 1 on b to rotate the state of a to 0
qc.cx(a,b) # uses the 1 on a to rotate the state of b to 0
qc.draw()
```

```python
qc.save_unitary()
qobj = assemble(qc)
uni = sim.run(qc).result().get_unitary()
array_to_latex(uni)
```

## 3. Controlled Rotations

```python
qc = QuantumCircuit(2)
theta = pi # theta can be anything (pi chosen arbitrarily)
qc.x(0)
qc.ry(theta/4,t)
qc.cx(c,t)
qc.ry(-theta/4,t)
qc.cx(c,t)
qc.draw()
```

```python
qc.save_statevector()

qobj = assemble(qc)
final_state = sim.run(qc).result().get_statevector()
plot_bloch_multivector(final_state)
```

```python
A = Gate('A', 1, [])
B = Gate('B', 1, [])
C = Gate('C', 1, [])
alpha = 1 # arbitrarily define alpha to allow drawing of circuit
```

```python
qc = QuantumCircuit(2)
qc.append(C, [t])
qc.cz(c,t)
qc.append(B, [t])
qc.cz(c,t)
qc.append(A, [t])
qc.p(alpha,c)
qc.draw()
```

## 4. Toffoli gate

```python
qc = QuantumCircuit(3)
a = 0
b = 1
t = 2
# Toffoli with control qubits a and b and target t
qc.ccx(a,b,t)
qc.draw()
```

```python
qc = QuantumCircuit(3)
qc.cp(theta,b,t)
qc.cx(a,b)
qc.cp(-theta,b,t)
qc.cx(a,b)
qc.cp(theta,a,t)
qc.draw()
```

```python
qc = QuantumCircuit(3)
qc.ch(a,t)
qc.cz(b,t)
qc.ch(a,t)
qc.draw()
```

- T 게이트는 Z축에 대해서 $\pi/4$만큼 회전하는 것이라 볼 수 있다.
$$R_{z}(\pi/4)=e^{i\pi/8Z}$$
- 이미 알려진 대로, $HZH=X$라 알려져 있다. 그리고 Z 게이트는 $\phi=\pi$일 때이다. 그렇다면 $\phi=\pi/4$인 T게이트에 대해서도 비슷한 공식을 적용했을 때 비슷한 효과를 기대할 수 있지 않을까?

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.t(0)
qc.h(0)
qc.draw()
```

위 결과는 다음의 회전에 대응한다.
$$R_z(\pi/4)R_x(\pi/4)$$

```python
qc.save_statevector()

qobj = assemble(qc)
final_state = sim.run(qc).result().get_statevector()
plot_bloch_multivector(final_state)
```
