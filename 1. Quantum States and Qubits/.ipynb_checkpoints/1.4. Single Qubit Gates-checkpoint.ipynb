```python
from qiskit import QuantumCircuit,assemble, Aer
from math import pi, sqrt
from qiskit.visualization import plot_bloch_multivector, plot_histogram
sim = Aer.get_backend('aer_simulator')
```

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.draw()
```

```python
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```

```python
def x_measurement(qc, qubit, cbit):
    qc.h(qubit)
    qc.measure(qubit, cbit)
    return qc
```

```python
initial_state = [1/sqrt(2), -1/sqrt(2)] # |->
qc = QuantumCircuit(1,1)
qc.initialize(initial_state, 0)
x_measurement(qc, 0, 0)
qc.draw()
```

Qiskit은 Computational basis(Z-basis)로만 측정을 지원한다. 따라서 다른 basis로 측정하기 위해서는 측정 전에 직접 작업을 해줘야한다.

```python
qobj = assemble(qc)
counts = sim.run(qobj).result().get_counts()
plot_histogram(counts)
```

## Quick Exercise

```python
init1 = [1/sqrt(2), 1/sqrt(2)]
init2 = [1/sqrt(2), -1/sqrt(2)]
qc1 = QuantumCircuit(1)
qc1.initialize(init1)
qc1.measure_all()
qc2 = QuantumCircuit(1)
qc1.initialize(init2)
qc2.measure_all()

qobj1 = assemble(qc1)
result1 = sim.run(qobj1).result().get_counts()
qobj2 = assemble(qc1)
result2 = sim.run(qobj2).result().get_counts()
```

```python
plot_histogram(result1)
```

```python
plot_histogram(result2)
```

```python
def y_measurement(qc, qubit, cbit):
    qc.sdg(qubit)
    qc.h(qubit)
    qc.measure(qubit, cbit)
    return qc
```

## Y-basis로 측정하기
입력이 Y-basis로 된 큐비트 상태가 있다고 가정하자.

```python
initial_state = [1/sqrt(2), 1j/sqrt(2)] # |+i>
qc = QuantumCircuit(1,1)
qc.initialize(initial_state, 0)
y_measurement(qc, 0, 0)
qc.draw()
```

```python
qobj = assemble(qc)
result = sim.run(qobj).result().get_counts()
plot_histogram(result)
```

```python
qc = QuantumCircuit(1)
qc.p(pi/4, 0)
qc.draw()
```

```python
qc = QuantumCircuit(1)
qc.s(0)   # Apply S-gate to qubit 0
qc.sdg(0) # Apply Sdg-gate to qubit 0
qc.draw()
```

```python
qc = QuantumCircuit(1)
qc.t(0)   # Apply T-gate to qubit 0
qc.tdg(0) # Apply Tdg-gate to qubit 0
qc.draw()
```

```python
qc = QuantumCircuit(1)
qc.u(pi/2, 0, pi, 0)
qc.draw()
```

```python
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```
