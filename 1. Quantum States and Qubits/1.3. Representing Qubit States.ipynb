```python
from qiskit import QuantumCircuit, assemble, Aer
from qiskit.visualization import plot_histogram,plot_bloch_vector
from math import sqrt, pi
sim = Aer.get_backend('aer_simulator')
```

```python
qc = QuantumCircuit(1)
initial_state = [0,1] # amplitudes of each state, |0>, |1>
qc.initialize(initial_state, 0)
qc.draw()
```

```python
qc.save_statevector()
qobj = assemble(qc)
result = sim.run(qobj).result()
```

```python
qc.measure_all()
qc.draw()
```

```python
counts = result.get_counts()
plot_histogram(counts)
```

```python
initial_state = [1/sqrt(2), 1j/sqrt(2)]

qc = QuantumCircuit(1)
qc.initialize(initial_state, 0) # init 0th qubit to 'initial_state'
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
print(state)
```

```python
qobj = assemble(qc)
results = sim.run(qobj).result().get_counts()
plot_histogram(results)
```

## Quick Exercise
1. Create a state vector that will give a 1/3 probability of measuring |0>

```python
qc = QuantumCircuit(1)
init_state = [1/sqrt(3), sqrt(2)/sqrt(3)]
qc.initialize(init_state, 0)
qc.save_statevector()
qobj = assemble(qc)
results = sim.run(qobj).result().get_counts()
plot_histogram(results)
```

2. Create a different state vector that will give the same measurement probabilities

```python
qc1 = QuantumCircuit(1)
qc2 = QuantumCircuit(1)
init_state1 = [1j/sqrt(3), sqrt(2)/sqrt(3)]
init_state2 = [1/sqrt(3), -sqrt(2)/sqrt(3)]
qc1.initialize(init_state1, 0)
qc1.save_statevector()
qc2.initialize(init_state2, 0)
qc2.save_statevector()

qobj1 = assemble(qc1)
qobj2 = assemble(qc2)
results1 = sim.run(qobj1).result().get_counts()
results2 = sim.run(qobj2).result().get_counts()
plot_histogram(results1)
```

```python
plot_histogram(results2)
```

```python
qc = QuantumCircuit(1) # We are redefining qc
initial_state = [0.+1.j/sqrt(2), 1/sqrt(2)+0.j]
qc.initialize(initial_state, 0)
qc.measure_all()
qc.save_statevector()
qc.draw()
```
