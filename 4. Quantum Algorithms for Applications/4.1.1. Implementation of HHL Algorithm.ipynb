```python
#initialization
import matplotlib.pyplot as plt
import numpy as np

# importing Qiskit
from qiskit import *

# import basic plot tools
from qiskit.visualization import plot_histogram
```

```python
aux = QuantumRegister(1, 'auxiliary')
phs = QuantumRegister(2, 'phase')
b = QuantumRegister(1, 'b')
meas = ClassicalRegister(2, 'measure')

hll = QuantumCircuit(aux, phs, b, meas)
hll.draw()
```

```python
# Step 1. initialize b
b_vec = [0,1]
hll.initialize(b_vec,3)

hll.barrier()
hll.draw()
```

```python
hll.h([1,2])
hll.draw()
```

```python
hll.cu(np.pi/2, -np.pi/2, np.pi/2, 3*np.pi/4,1,3)
#hll.cu(np.pi/2, -np.pi/2, np.pi/2, 0,1,3)
hll.cu(np.pi, np.pi, 0, 0, 2,3)
hll.draw()
```

```python
hll.barrier()
hll.h(2)
hll.cp(-np.pi/2,1,2)
hll.h(1)
hll.swap(1,2)
hll.draw()
```

```python
hll.barrier()
hll.cry(np.pi, 1,0)
hll.cry(np.pi/3, 2,0)
hll.draw()
```

```python
hll.barrier()
hll.draw()
```

```python
hll.barrier()
hll.swap(1,2)
hll.h(1)
hll.cp(np.pi/2,1,2)
hll.h(2)
hll.draw()
```

```python
hll.barrier()
hll.cu(np.pi, np.pi, 0, 0, 2,3)
hll.cu(np.pi/2, np.pi/2, -np.pi/2, -3*np.pi/4,1,3)
#hll.cu(np.pi/2, np.pi/2, -np.pi/2, 0,1,3)
hll.draw()
```

```python
hll.h([1,2])
hll.barrier()

hll.measure(0, 0)
hll.measure(3,1)

hll.draw()
```

```python
aer_sim = Aer.get_backend('aer_simulator')
shots = 2048
t_hll = transpile(hll, aer_sim)
qobj = assemble(t_hll, shots=shots)
results = aer_sim.run(qobj).result()
answer = results.get_counts()

plot_histogram(answer)
```
